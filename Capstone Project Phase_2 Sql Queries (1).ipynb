Table_1

1)  Write a SQL query to order records by a rental price  column in ascending order.

```python
mycursor.execute("SELECT Rental_price FROM table_1 order by Rental_price Desc")
myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

2)    Write a SQL query  to select unique combinations of City and State with their average Rental Price.

```python
mycursor.execute("SELECT City,State_Code,avg(Rental_price) as Avg  FROM table_1 group by City")
myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

3.Write a SQL query to select the top 5 highest deposit amounts with corresponding Address and City .

```python
mycursor.execute("SELECT City,Address,Deposit from table_1 order by Deposit Asc limit 5  ")
myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

4. Write a SQL query to select the count of records for each Country along with the total deposit amount.

```python
mycursor.execute("SELECT Count(Country) from table_1 where Deposit is NOT NUll  ")
myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

5.Write a SQL query to select records with a Rental Price higher than the average Rental Price across all records.

```python
mycursor.execute("SELECT Count(Rental_price) from table_1  where Rental_price > (SELECT round(avg(Rental_price),2)  from table_1)")
myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```
